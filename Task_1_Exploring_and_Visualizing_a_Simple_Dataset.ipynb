```python
# imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the built-in Iris dataset from seaborn
iris = pd.read_csv('Iris.csv')
```

```python
# Shape of the dataset
print("Shape of dataset:", iris.shape)

# Column names
print("\nColumn names:", iris.columns.tolist())

# First five rows
print("\nFirst five rows:")
print(iris.head())

# Summary of dataset
print("\nDataset info:")
print(iris.info())

# Descriptive statistics
print("\nDescriptive statistics:")
print(iris.describe())
```

```python

sns.scatterplot(data=iris, x="SepalLengthCm", y="SepalWidthCm", hue="Species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
iris.hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Features")
plt.show()
```
